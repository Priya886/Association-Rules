```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
movies = pd.read_csv('I:\Assigment - Data Science\Association Rules/my_movies.csv')
```

```python
movies
```

```python
te = TransactionEncoder()
te_ary = te.fit(movies).transform(movies)
df = pd.Dataframe(te_ary,columns = te.columns_)
df
```

```python
df=pd.get_dummies(movies)
df.head()
```

```python
frequent_itemsets = apriori(df, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
frequency_itemsets = apriori(df,min_support = 0.3,use_colnames = True)
frequency_itemsets['length'] = frequency_itemsets['itemsets'].apply(lambda x : len(x))
frequency_itemsets
```

```python
rules = association_rules(frequent_itemsets,min_threshold = 1)
rules.head()
```

```python
rules [(rules['lift']>= 1.6)
       (rules['confidence'] > = 1.0)]
```

```python
rules[rules.lift>2]
```

```python
rules[rules.confidence>0.8]
```
